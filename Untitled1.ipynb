```python
import geopandas as gpd
import pandas as pd
```

```python
gadm = gpd.read_file('data/gadm41/gadm41_PHL_1.shp')
gadm.head()
```

```python
gadm.shape
```

```python
gadm.NAME_1
```

```python
region_prov = pd.read_csv('data/prov_region.csv')
region_prov.head()
```

```python
gadm_region = gadm.merge(region_prov, left_on='NAME_1', right_on='province', how='left')
gadm_region
```

```python
gadm_region_disoolved = gadm_region.dissolve(by='region')
gadm_region_disoolved.shape
```

```python
gadm_region_disoolved.head()
```

```python
regions = gadm_region_disoolved[['geometry']].reset_index()
regions.head()
```

```python
regions.to_file('gadm_regions.geojson', driver='GeoJSON')
```
